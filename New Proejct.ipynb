```python
import gradio as gr
from groq import Groq

# Groq client
client = Groq(api_key="Learn_App")

def generate_roadmap(name, field):

    prompt = f"Create a step-by-step learning roadmap for {name} who wants to learn {field}. Include beginner to advanced steps."

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "system", "content": "You are an expert AI career mentor."},
            {"role": "user", "content": prompt}
        ]
    )

    roadmap = response.choices[0].message.content
    return roadmap


app = gr.Interface(
    fn=generate_roadmap,
    inputs=[
        gr.Textbox(label="Your Name"),
        gr.Textbox(label="Field you want to learn (AI, Data Science, Web Dev etc)")
    ],
    outputs=gr.Textbox(label="Your AI Learning Roadmap"),
    title="AI Learning Roadmap Generator",
    description="Enter your name and field to generate a personalized AI learning roadmap."
)

app.launch()
```
